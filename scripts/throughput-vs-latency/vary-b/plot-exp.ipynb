```python
# Build a ping percentile plot

import matplotlib.pyplot as plt
from numpy import percentile
from csv import reader
import math
```

```python
shs_w = []
shs_lat = []
shs_tput = []

apl_w = []
apl_lat = []
apl_tput = []

with open("synchs.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_w.append(float(row[0])/1000)
        shs_tput.append(float(row[1])/1000)
        shs_lat.append(float(row[2]))

with open("apollo.csv") as f:
    rdr = reader(f)
    for row in rdr:
        apl_w.append(float(row[0])/1000)
        apl_tput.append(float(row[1])/1000)
        apl_lat.append(float(row[2]))
```

```python
plt.rc('text',usetex=True)
plt.plot(shs_tput,shs_lat, label="Sync HotStuff (b100)", linestyle='--', marker='o')
plt.plot(apl_tput,apl_lat, label="Apollo (b100)", linestyle='--', marker='+')
plt.legend()
plt.xlabel("Throughput (in Kops/s)")
plt.ylabel("Latency (in ms)")
# plt.show()
# plt.xticks(log_points)
plt.savefig("Tput-vs-latency-for-diff-b.pdf")
```

```python
plt.rc('text',usetex=True)
plt.plot(shs_w,shs_lat, label="Sync HotStuff (b100)", linestyle='--', marker='o')
plt.plot(apl_w,apl_lat, label="Apollo (b100)", linestyle='--', marker='+')
plt.legend()
plt.xlabel("Offered Load $w$ (in Kops)")
plt.ylabel("Latency (in ms)")
# plt.show()
# plt.xticks(log_points)
plt.savefig("Latency-vs-w-for-diff-b.pdf")
```

```python
plt.rc('text',usetex=True)
plt.plot(shs_w,shs_tput, label="Sync HotStuff (b100)", linestyle='--', marker='o')
plt.plot(apl_w,apl_tput, label="Apollo (b100)", linestyle='--', marker='+')
plt.legend()
plt.xlabel("Offered Load $w$ (in Kops)")
plt.ylabel("Throughput (in Kops/s)")
# plt.show()
# plt.xticks(log_points)
plt.savefig("Tput-vs-w-for-diff-b.pdf")
```
