```python
# Build a ping percentile plot

import matplotlib.pyplot as plt
from numpy import percentile
from csv import reader
import math
```

```python
shs_b100_w = []
shs_b100_lat = []
shs_b100_tput = []
apl_b100_w = []
apl_b100_lat = []
apl_b100_tput = []

shs_b400_w = []
shs_b400_lat = []
shs_b400_tput = []
apl_b400_w = []
apl_b400_lat = []
apl_b400_tput = []

shs_b800_w = []
shs_b800_lat = []
shs_b800_tput = []
apl_b800_w = []
apl_b800_lat = []
apl_b800_tput = []

with open("../../../c5-exp-varyb/b100-synchs-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_b100_w.append(float(row[0])/1000)
        shs_b100_tput.append(float(row[1])/1000)
        shs_b100_lat.append(float(row[2]))

with open("../../../c5-exp-varyb/b400-synchs-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_b400_w.append(float(row[0])/1000)
        shs_b400_tput.append(float(row[1])/1000)
        shs_b400_lat.append(float(row[2]))

with open("../../../c5-exp-varyb/b800-synchs-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_b800_w.append(float(row[0])/1000)
        shs_b800_tput.append(float(row[1])/1000)
        shs_b800_lat.append(float(row[2]))

with open("../../../c5-exp-varyb/b100-apollo-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        apl_b100_w.append(float(row[0])/1000)
        apl_b100_tput.append(float(row[1])/1000)
        apl_b100_lat.append(float(row[2]))

with open("../../../c5-exp-varyb/b400-apollo-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        apl_b400_w.append(float(row[0])/1000)
        apl_b400_tput.append(float(row[1])/1000)
        apl_b400_lat.append(float(row[2]))

with open("../../../c5-exp-varyb/b800-apollo-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        apl_b800_w.append(float(row[0])/1000)
        apl_b800_tput.append(float(row[1])/1000)
        apl_b800_lat.append(float(row[2]))
```

```python
plt.rc('text',usetex=True)

plt.plot(shs_b100_tput,shs_b100_lat, label="Sync HotStuff (b100)", linestyle='--', marker='o')
plt.plot(apl_b100_tput,apl_b100_lat, label="Apollo (b100)", linestyle='--', marker='+')

plt.plot(shs_b400_tput,shs_b400_lat, label="Sync HotStuff (b400)", linestyle='--', marker='o')
plt.plot(apl_b400_tput,apl_b400_lat, label="Apollo (b400)", linestyle='--', marker='+')

plt.plot(shs_b800_tput,shs_b800_lat, label="Sync HotStuff (b800)", linestyle='--', marker='o')
plt.plot(apl_b800_tput,apl_b800_lat, label="Apollo (b800)", linestyle='--', marker='+')

plt.legend()
plt.xlabel("Response Rate (in Kops/s)")
plt.ylabel("Latency (in ms)")
# plt.show()
# plt.xticks(log_points)
plt.savefig("Tput-vs-latency-for-diff-b.pdf")
```

```python
plt.rc('text',usetex=True)

plt.plot(shs_b100_w,shs_b100_lat, label="Sync HotStuff (b100)", linestyle='--', marker='o')
plt.plot(apl_b100_w,apl_b100_lat, label="Apollo (b100)", linestyle='--', marker='+')

plt.plot(shs_b400_w,shs_b400_lat, label="Sync HotStuff (b400)", linestyle='--', marker='o')
plt.plot(apl_b400_w,apl_b400_lat, label="Apollo (b400)", linestyle='--', marker='+')

plt.plot(shs_b800_w,shs_b800_lat, label="Sync HotStuff (b800)", linestyle='--', marker='o')
plt.plot(apl_b800_w,apl_b800_lat, label="Apollo (b800)", linestyle='--', marker='+')

plt.legend()
plt.xlabel("Offered Load $w$ (in Kops)")
plt.ylabel("Latency (in ms)")
# plt.show()
# plt.xticks(log_points)
plt.savefig("Latency-vs-w-for-diff-b.pdf")
```
